```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
sns.set_style("white")
```

1. 📈 Normal Distribution (Gaussian)

Theoretical (SciPy)

```python
x = np.linspace(-5, 5, 1000)
pdf = stats.norm.pdf(x, loc=0, scale=1)  # mean=0, std=1
```

Practical (NumPy)

```python
samples = np.random.normal(loc=0, scale=1, size=100000)
```

Plot

```python
x = np.linspace(-5, 5, 1000)
plt.figure(figsize=(14, 4))

# PDF
plt.subplot(1, 3, 1)
plt.plot(x, stats.norm.pdf(x, loc=0, scale=1), label='μ=0,σ=1', color='blue')
plt.plot(x, stats.norm.pdf(x, loc=0, scale=2), label='μ=0,σ=2', color='red')
plt.plot(x, stats.norm.pdf(x, loc=2, scale=1), label='μ=2,σ=1', color='green')
plt.legend()
plt.title("Normal - PDF")
plt.grid()

# CDF
plt.subplot(1, 3, 2)
plt.plot(x, stats.norm.cdf(x, loc=0, scale=1), label='μ=0,σ=1', color='blue')
plt.plot(x, stats.norm.cdf(x, loc=0, scale=2), label='μ=0,σ=2', color='red')
plt.plot(x, stats.norm.cdf(x, loc=2, scale=1), label='μ=2,σ=1', color='green')
plt.legend()
plt.title("Normal - CDF")
plt.grid()

# Histogram
plt.subplot(1, 3, 3)
plt.hist(np.random.normal(loc=0, scale=1, size=1000), bins=30, density=True, label='μ=0,σ=1', color='blue')
plt.hist(np.random.normal(loc=0, scale=2, size=1000), bins=30, density=True, alpha=0.7, label='μ=0,σ=2', color='red')
plt.hist(np.random.normal(loc=2, scale=1, size=1000), bins=30, density=True, alpha=0.7, label='μ=2,σ=1', color='green')
plt.legend()
plt.title("Normal - Practical Histogram")
plt.grid()

plt.tight_layout()
plt.suptitle('Normal Distribution', y=1.1, fontsize=20)
plt.show()
```

2. 🎲 Uniform Discrete Distribution

Theoretical (SciPy)

```python
x = np.arange(1, 7)  # Like rolling a dice
pmf = stats.randint.pmf(x, low=1, high=7)
```

Practical (NumPy)

```python
samples = np.random.randint(1, 7, size=1000)
```

Plot

```python
x = np.arange(1, 7)
pmf = stats.randint.pmf(x, 1, 7)
cdf = stats.randint.cdf(x, 1, 7)
samples = np.random.randint(1, 7, size=1000)

plt.figure(figsize=(14, 4))

# PMF
plt.subplot(1, 3, 1)
plt.stem(x, pmf, basefmt=" ")
plt.title("Uniform Discrete - PMF")
plt.grid()

# CDF
plt.subplot(1, 3, 2)
plt.step(x, cdf, where='mid', color='red')
plt.title("Uniform Discrete - CDF")
plt.grid()

# Histogram
plt.subplot(1, 3, 3)
plt.hist(samples, bins=np.arange(1, 8)-0.5, rwidth=0.8, density=True, color="orange")
plt.title("Uniform Discrete - Practical Histogram")
plt.grid()

plt.tight_layout()
plt.suptitle('Uniform Discrete Distribution', y=1.1, fontsize=20)
plt.show()
```

7 ⏳ Exponential Distribution (Continuous)

🧪 Practical (NumPy)

```python
x = np.linspace(0, 10, 1000)
plt.figure(figsize=(14, 4))

# PDF
plt.subplot(1, 3, 1)
plt.plot(x, stats.expon.pdf(x,1), label='λ = 1', color='blue')
plt.plot(x,stats.expon.pdf(x,1/3),label='λ = 3',color='red')
plt.plot(x,stats.expon.pdf(x,1/10),label='λ = 10',color='green')
plt.legend()
plt.title("Exponential - PDF")
plt.grid()

# CDF
plt.subplot(1, 3, 2)
plt.plot(x, stats.expon.cdf(x,1), label='λ = 1', color='blue')
plt.plot(x,stats.expon.cdf(x,1/3),label='λ = 3',color='red')
plt.plot(x,stats.expon.cdf(x,1/10),label='λ = 10',color='green')
plt.title("Exponential - CDF")
plt.legend()
plt.grid()

# Histogram
plt.subplot(1, 3, 3)
plt.hist(np.random.exponential(scale=1, size=1000), bins=30, density=True, label='λ = 1', color='blue')
plt.hist(np.random.exponential(scale=1/3, size=1000), bins=30, density=True, alpha=0.7, label='λ = 3', color='red')
plt.hist(np.random.exponential(scale=1/10, size=1000), bins=30, density=True, alpha=0.5, label='λ = 10', color='green')
plt.title("Exponential - Practical Histogram")
plt.legend()
plt.grid()

plt.tight_layout()
plt.suptitle('Exponential Distribution', y=1.1, fontsize=20)
plt.show()
```

9) t-Distribution

🎓 Theoretical (SciPy)

```python
x = np.linspace(-5, 5, 1000)
pdf = stats.t.pdf(x, df=1)
cdf = stats.t.cdf(x, df=1)
```

🧪 Practical (NumPy)

```python
samples = np.random.standard_t(df=1, size=100)
```

📊 Plot


```python
x = np.linspace(-5, 5, 1000)
plt.figure(figsize=(15, 5))

# PDF
plt.subplot(1, 3, 1)
plt.plot(x, stats.t.pdf(x, df=1), label='df = 1', color='blue')     # Heavy tails
plt.plot(x, stats.t.pdf(x, df=5), label='df = 5', color='red')      # Moderate
plt.plot(x, stats.t.pdf(x, df=10), label='df = 10', color='green')  # Close to normal
plt.plot(x, stats.norm.pdf(x), label='Normal Distribution', color='black')  # Close to normal
plt.legend()
plt.title("t-Distribution - PDF")
plt.grid()

# CDF
plt.subplot(1, 3, 2)
plt.plot(x, stats.t.cdf(x, df=1), label='df = 1', color='blue')
plt.plot(x, stats.t.cdf(x, df=5), label='df = 5', color='red')
plt.plot(x, stats.t.cdf(x, df=10), label='df = 10', color='green')
plt.plot(x, stats.norm.cdf(x), label='Normal Distribution', color='k')  # Close to normal
plt.legend()
plt.title("t-Distribution - CDF")
plt.grid()

# Histogram (sampled data)
plt.subplot(1, 3, 3)
plt.hist(np.random.standard_t(df=1, size=100), bins=30, density=True, label='df = 1', color='blue')
plt.hist(np.random.standard_t(df=5, size=100), bins=30, density=True, label='df = 5', color='red', alpha=0.7)
plt.hist(np.random.standard_t(df=10, size=100), bins=30, density=True, label='df = 10', color='green', alpha=0.5)
plt.hist(np.random.normal(size=100), label='Normal Distribution',  bins=30, density=True, color='k', alpha=0.3)  # Close to normal
plt.xlim(-5,5)
plt.legend()
plt.title("t-Distribution - Practical Histogram")
plt.grid()

plt.tight_layout()
plt.suptitle('t-Distribution', y=1.1, fontsize=20)
plt.show()
```

310- Central Limit Theorem

```python
plt.figure(figsize=(15, 5))
def central_limit_theorem(λ, size_of_sample, number_of_samples):
    means = []
    for i in range(number_of_samples):
        means.append(np.random.exponential(1/λ, size=size_of_sample).mean())
    return means

plt.subplot(1, 2, 1)
sns.histplot(central_limit_theorem(1, 20, 1000), bins=30, kde=True, color='blue', label='λ = 1 | n = 20', alpha=1)
sns.histplot(central_limit_theorem(2, 20, 1000), bins=30, kde=True, color='red', label='λ = 2 | n = 20', alpha=0.8)
sns.histplot(central_limit_theorem(3, 20, 1000), bins=30, kde=True, color='green', label='λ = 3 | n = 20', alpha=0.5)
plt.legend()
text = "\n".join([
    f"λ=1 → Population Mean: {1/1:.0f} | Sample Mean: {np.mean(central_limit_theorem(1, 20, 1000)):.4f}",
    f"λ=2 → Population Mean: {1/2:.1f} | Sample Mean: {np.mean(central_limit_theorem(2, 20, 1000)):.4f}",
    f"λ=3 → Population Mean: {1/3:.1f} | Sample Mean: {np.mean(central_limit_theorem(3, 20, 1000)):.4f}"
])
plt.text(0.1, -0.2, text, va='center', transform=plt.gca().transAxes, fontsize=12)
plt.grid()

plt.subplot(1, 2, 2)
sns.histplot(central_limit_theorem(2, 10, 1000), bins=30, kde=True, color='blue', label='λ = 2 | n = 10', alpha=1)
sns.histplot(central_limit_theorem(2, 100, 1000), bins=30, kde=True, color='red', label='λ = 2 | n = 100', alpha=0.8)
sns.histplot(central_limit_theorem(2, 1000, 1000), bins=30, kde=True, color='green', label='λ = 2 | n = 1000', alpha=0.6)
plt.legend()
text = "\n".join([
    f"n=10 → Population Mean: {1/2:.1f} | Sample Mean: {np.mean(central_limit_theorem(2, 10, 1000)):.4f}",
    f"n=100 → Population Mean: {1/2:.1f} | Sample Mean: {np.mean(central_limit_theorem(2, 100, 1000)):.4f}",
    f"n=1000 → Population Mean: {1/2:.1f} | Sample Mean: {np.mean(central_limit_theorem(2, 1000, 1000)):.4f}","",
    "                              n ↑ SE ↓"
])
plt.text(0.15, -0.25, text,va='center', transform=plt.gca().transAxes, fontsize=12)
plt.grid()

plt.tight_layout()
plt.suptitle('Central Limit Theorem', y=1.1, fontsize=25)
plt.show()
```
